```python
import pandas as pd
import ast

# Ruta del archivo original
file_path ="C:\\Users\\yanin\\OneDrive\\Desktop\\proyecto 1\\archivos\\credits (2).csv"

# Cargar el archivo CSV
df = pd.read_csv(file_path)

# Tomar una muestra aleatoria de 1000 filas
df_reduced = df.sample(n=1000, random_state=42)

# Convertir las columnas 'cast' y 'crew' de string a listas de diccionarios
for column in ["cast", "crew"]:
    df_reduced[column] = df_reduced[column].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else [])

# Guardar el archivo reducido
reduced_file_path = "C:\\Users\\yanin\\Downloads\\archivo_reducido.csv"
df_reduced.to_csv(reduced_file_path, index=False)
print("\nâœ… Archivo reducido guardado en:", reduced_file_path)
print("\nðŸ“Š Resumen del archivo reducido:")
print(df_reduced.head())  # Muestra las primeras filas

# FunciÃ³n para desanidar una columna (cast o crew)
def desanidar(df, column_name):
    df_exploded = df.explode(column_name)
    df_exploded = df_exploded[df_exploded[column_name].notna()]
    df_exploded = df_exploded[column_name].apply(pd.Series)
    df_exploded.rename(columns={"id": f"{column_name}_id", "profile_path": "profile_part"}, inplace=True)
    df_exploded[f"{column_name}_id"] = df_exploded[f"{column_name}_id"].astype("Int64")
    return df_exploded

# Procesar 'cast' y 'crew'
df_cast = desanidar(df_reduced, "cast")
df_crew = desanidar(df_reduced, "crew")

# Guardar los archivos desanidados
cast_file_path = "C:\\Users\\yanin\\OneDrive\\Desktop\\proyecto 1\\archivos\\cast_desanidado.csv"

crew_file_path = "C:\\Users\\yanin\\OneDrive\\Desktop\\proyecto 1\\archivos\\crew_desanidado.csv"

df_cast.to_csv(cast_file_path, index=False)
df_crew.to_csv(crew_file_path, index=False)

print("\nâœ… Archivo 'cast' desanidado guardado en:", cast_file_path)
print("\nðŸ“Š Resumen del archivo 'cast' desanidado:")
print(df_cast.head())  # Muestra las primeras filas

print("\nâœ… Archivo 'crew' desanidado guardado en:", crew_file_path)
print("\nðŸ“Š Resumen del archivo 'crew' desanidado:")
print(df_crew.head())  # Muestra las primeras filas
```
